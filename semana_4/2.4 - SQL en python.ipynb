```python
# cargar datos

import pandas as pd

df=pd.read_csv('data/shoes.csv')

df.head()
```

```python
# seleccion de datos

df=df[['brand', 'colors', 'sizes', 'prices.amountMax', 'imageURLs', 'ean', 'upc']]
df.head()
```

```python
# quita nan

df=df.dropna().reset_index(drop=True)

df.head()
```

```python
df.info(memory_usage='deep')
```

```python
df.sizes=df.sizes.apply(lambda x: x.split(','))
df.colors=df.colors.apply(lambda x: x.split(','))

df.head()
```

```python
# un registro por cada talla y cada color

df=df.explode('sizes').reset_index(drop=True)
df=df.explode('colors').reset_index(drop=True)

df.head()
```

```python
df.info(memory_usage='deep')
```

```python
import warnings
warnings.simplefilter('ignore')

codigo=pd.read_csv('data/upc_corpus.csv')
codigo=codigo.dropna().reset_index(drop=True)

codigo.head()
```

```python
codigo.shape
```

```python
df.ean=codigo.ean.iloc[0:len(df)]
```

```python
# renombrar

df=df.rename(columns={'colors':'color', 
                      'sizes':'size', 
                      'prices.amountMax':'price', 
                      'imageURLs':'url'})

df.head()
```

```python
df['id']=[i for i in range(len(df))]

df=df.drop(columns=['upc'])

df.head()
```

```python
datos=pd.read_sql('''select prime_genre,
                            sum(rating_count_tot) as total_rating,
                            avg(user_rating) as avg_rating
                     from Ratings
                     where price=0
                     group by prime_genre
                     order by total_rating desc
                     limit 10;''', motor)
datos.head()
```

```python
str_conn='mysql+pymysql://root:password@localhost:3306/publications'

motor=create_engine(str_conn)
```

```python
query='''
        select group_concat(t.title) as titles,
        t.year
        from (select title, year(max(pubdate)) as 'year',
        count(titleauthor.title_id) as num_titles,
        group_concat(concat(authors.au_fname, ' ', authors.au_lname)) as authors
        from titles
        left join titleauthor
        on titles.title_id=titleauthor.title_id
        left join authors
        on authors.au_id=titleauthor.au_id
        group by titles.title
        order by num_titles desc) as t
        group by t.year;

'''
```

```python
df=pd.read_sql(query, motor)

df.head()
```
